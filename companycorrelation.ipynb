```python
ContractName = 'RB.SHF'
StartDate = '2015-01-01'
EndDate = '2018-12-31'
Name_Date = Get_Contracts(mySqlDBReader, ContractName, StartDate, EndDate)
L = []
for i in range(len(Name_Date)):
    contract = Name_Date[i][2]
    start = Name_Date[i][0]
    end = Name_Date[i][1]
    correlationTable, corr = CorrelationTable(mySqlDBLocal, '持买仓量', Companys, contract, start, end)
    L.append(corr)
```

```python
ans = MakeCorrTable(L)
```

```python
ans2 = ans.applymap(lambda x:float("%.2f" %x))
ans2.fillna('-')
temp = ans2.iloc[:,list(range(0 , 10,3))]
temp['均值'] = temp.mean(axis = 1)
temp['计数'] = temp.count(axis = 1)-1
temp.dropna(subset=['均值']).fillna('-').sort_values(['计数','均值'],ascending = False)
```
